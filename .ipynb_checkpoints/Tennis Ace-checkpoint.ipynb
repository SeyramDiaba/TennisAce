## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
pd.set_option('display.max_columns', None)
```

## Load and Inspect Dataset

```python
# load and investigate the data here:
df = pd.read_csv('tennis_stats.csv')
df.head()
```

# Exploratory Data Analysis

```python
# perform exploratory analysis here:
offensive = df[['Aces','DoubleFaults','FirstServe','FirstServePointsWon','SecondServePointsWon','BreakPointsFaced','BreakPointsSaved','ServiceGamesPlayed','ServiceGamesWon','TotalServicePointsWon']]
defensive = df[['FirstServeReturnPointsWon','SecondServeReturnPointsWon','BreakPointsOpportunities','BreakPointsConverted','ReturnGamesPlayed','ReturnGamesWon','ReturnPointsWon','TotalPointsWon']]
features = offensive + defensive
outcomes = df[['Wins','Losses','Winnings','Ranking']]
```

##  Offensive Feature plot against outcomes

```python
# plot of features against win
# offensive
# Wins
def feat_out1(offensive,outcomes):
    for col in offensive.columns:
        plt.scatter(outcomes['Wins'],offensive[col], alpha= 0.4)
        plt.xlabel('Wins')
        plt.ylabel(col)
        plt.title(col +' against Wins Scatter Plot')
        plt.show()
        plt.clf()
        # Losses
    for col in offensive.columns:
        plt.scatter(outcomes['Losses'], offensive[col], alpha = 0.4)
        plt.xlabel('Losses')
        plt.ylabel(col)
        plt.title(col + ' against Losses Scatter Plot')
        plt.show()
        plt.clf()
        
        # Winnings
    for col in offensive.columns:
        plt.scatter(outcomes['Winnings'], offensive[col], alpha = 0.4)
        plt.xlabel('Winnings')
        plt.ylabel(col)
        plt.title(col + ' against Winnings Scatter Plot')
        plt.show()
        plt.clf()
        
        # Ranking
    for col in offensive.columns:
        plt.scatter(outcomes['Ranking'], offensive[col], alpha = 0.4)
        plt.xlabel('Ranking')
        plt.ylabel(col)
        plt.title(col + ' against Ranking Scatter Plot')
        plt.show()
        plt.clf()
        
```

```python
feat_out1(offensive,outcomes)
```

## Defensive feature plot against outcomes

```python
# Defensive
# Wins
def feat_out2(defensive,outcomes):
    for col in offensive.columns:
        plt.scatter(outcomes['Wins'],offensive[col], alpha= 0.4)
        plt.xlabel('Wins')
        plt.ylabel(col)
        plt.title(col +' against Wins Scatter Plot')
        plt.show()
        plt.clf()
        
        # Losses
    for col in defensive.columns:
        plt.scatter(outcomes['Losses'], defensive[col], alpha = 0.4)
        plt.xlabel('Losses')
        plt.ylabel(col)
        plt.title(col + ' against Losses Scatter Plot')
        plt.show()
        plt.clf()
        
        # Winnings
    for col in defensive.columns:
        plt.scatter(outcomes['Winnings'], defensive[col], alpha = 0.4)
        plt.xlabel('Winnings')
        plt.ylabel(col)
        plt.title(col + ' against Winnings Scatter Plot')
        plt.show()
        plt.clf()
        
        # Ranking
    for col in defensive.columns:
        plt.scatter(outcomes['Ranking'], defensive[col], alpha = 0.4)
        plt.xlabel('Ranking')
        plt.ylabel(col)
        plt.title(col + ' against Ranking Scatter Plot')
        plt.show()
        plt.clf()
        
        
```

```python
feat_out2(defensive,outcomes)
```

## Single feature to Predict Winnings

```python
def linear_regression(features,outcomes):
    X = df[features]
    y = df[outcomes]
    
    x_train,x_test,y_train,y_test = train_test_split(X,y,train_size =0.8,test_size=0.2, random_state = 1)
    model = LinearRegression()
    print("Train Score: ",model.fit(x_train,y_train))
    print("Test Score: ",model.score(x_test,y_test))
    y_predict = model.predict(x_test)
    coef = sorted(list(zip(features, model.coef_)), key = lambda x: abs(x[1]), reverse=True)
    for i in range(len(coef)):
        print(f'{coef[i][0]}: {coef[i][1]}')
    y_predicted = model.predict(x_test)
    plt.figure(figsize=(5, 5))
    plt.scatter(y_test, y_predicted, alpha=0.4)
    plt.xlabel(outcomes)
    plt.ylabel("Predicted " + outcomes)
    plt.show()
    
```

```python
linear_regression(['ServiceGamesPlayed'], 'Winnings')
```

```python
linear_regression(['BreakPointsConverted'],'Winnings')
```

```python
linear_regression(['FirstServePointsWon'],'Winnings')
```

## Two Features to Predict Winnings

```python
linear_regression(['BreakPointsOpportunities',
'FirstServeReturnPointsWon'],'Winnings')
```

```python
linear_regression(['ServiceGamesPlayed', 'ReturnGamesPlayed'], 'Winnings')
```

```python
linear_regression(['ServiceGamesPlayed', 'BreakPointsOpportunities'], 'Winnings')
```
